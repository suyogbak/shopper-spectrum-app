The rapid growth of e-commerce has resulted in the generation of massive volumes of customer transaction data. ## **DATA DICTIONARY**

1. **`InvoiceNo`** – Unique invoice number assigned to each transaction
   Represents a specific purchase transaction. Cancelled invoices are typically identified by invoice numbers starting with the letter "C".

2. **`StockCode`** – Unique product identifier
   Used to uniquely identify each product available in the inventory.

3. **`Description`** – Product description or name
   Provides a textual description of the purchased item and helps in understanding product categories and recommendation generation.

4. **`Quantity`** – Number of units purchased
   Indicates how many units of a product were included in a particular transaction.

5. **`InvoiceDate`** – Date and time of the transaction
   Captures when the purchase occurred and is essential for time-based analysis and RFM calculations.

6. **`UnitPrice`** – Price per unit of the product
   Represents the selling price of a single unit of the product before calculating the total transaction value.

7. **`CustomerID`** – Unique customer identifier
   Used to track customer purchase behavior across multiple transactions and serves as the primary key for customer-level analysis.

8. **`Country`** – Country where the customer is located
   Enables geographical analysis of customer behavior, purchasing trends, and market performance.

```python
import datetime as dt

# Calculate a snapshot date for recency calculation (one day after the last transaction in the dataset)
snapshot_date = data['InvoiceDate'].max() + dt.timedelta(days=1)

# Group by CustomerID to create the RFM table
rfm_table = data.groupby('CustomerID').agg({
    'InvoiceDate': lambda date: (snapshot_date - date.max()).days, # Recency
    'InvoiceNo': lambda num: num.nunique(),                       # Frequency
    'Sales Price': lambda price: price.sum()                      # Monetary
})

# Rename the columns for clarity
rfm_table.rename(columns={'InvoiceDate': 'Recency',
                          'InvoiceNo': 'Frequency',
                          'Sales Price': 'Monetary'},
                   inplace=True)

print("RFM Table (first 5 rows):")
print(rfm_table.head())
```

### **USING ELBOW METHOD TO FIND OPTIMAL NUMBER OF CLUSTERS**

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Standardize the RFM data
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_table)

wcss = []
# Try different numbers of clusters from 1 to 10
for i in range(1, 21):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init=10)
    kmeans.fit(rfm_scaled)
    wcss.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(10, 6))
plt.plot(range(1, 21), wcss, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal K (RFM)')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.grid(True)
plt.show()
```

### **BUILDING RECOMMENDATION SYSTEM USING COSINE SIMILARITY**

```python
# To build a product recommendation system, we'll use cosine similarity.
# First, we need to create a product-invoice matrix where rows are products (StockCode),
# columns are invoices (InvoiceNo), and values are the quantities purchased.

# Create a pivot table for product-invoice matrix
# Fill NaN values with 0, as it means the product was not part of that invoice.
product_invoice_matrix = data.pivot_table(
    index='StockCode',
    columns='InvoiceNo',
    values='Quantity'
).fillna(0)

# Import cosine_similarity from sklearn
from sklearn.metrics.pairwise import cosine_similarity

# Calculate the cosine similarity between products
# The result will be a square matrix where each element (i, j) represents
# the similarity between product i and product j.
product_similarity = cosine_similarity(product_invoice_matrix)

# Convert the similarity matrix into a DataFrame for easier handling and readability
# The index and columns will be the StockCodes of the products.
product_similarity_df = pd.DataFrame(
    product_similarity,
    index=product_invoice_matrix.index,
    columns=product_invoice_matrix.index
)

# Recommendation function
def recommend_products(stock_code, n_recommendations=5):
    """
    Recommends similar products based on cosine similarity.

    Args:
        stock_code (str): The StockCode of the product for which to find recommendations.
        n_recommendations (int): The number of top similar products to recommend.

    Returns:
        list: A list of dictionaries, where each dictionary contains 'StockCode' and 'Description'
              of the recommended products.
        str: An error message if the product is not found.
    """
    if stock_code not in product_similarity_df.index:
        return f"Product with StockCode '{stock_code}' not found in the similarity matrix."

    # Get the similarity scores for the given product
    # Sort them in descending order and get the top N, excluding the product itself.
    similar_scores = product_similarity_df[stock_code].sort_values(ascending=False)

    # Get the StockCodes of the most similar products
    # Use [1:n_recommendations+1] to exclude the product itself (which has a similarity of 1)
    # and get the next 'n_recommendations' products.
    similar_product_codes = similar_scores.index[1:n_recommendations+1]

    # Get the descriptions for the recommended products using the pre-existing map
    recommended_products_with_desc = []
    for code in similar_product_codes:
        # Ensure we handle cases where a StockCode might not have a description in the map
        # although 'stock_description_map' should cover most cases if created correctly earlier.
        if code in stock_description_map.index:
            desc = stock_description_map.loc[code, 'Description']
            recommended_products_with_desc.append({'StockCode': code, 'Description': desc})
        else:
            recommended_products_with_desc.append({'StockCode': code, 'Description': 'Description Not Found'})

    return recommended_products_with_desc

print("Product recommendation function `recommend_products(stock_code, n_recommendations=5)` created.")
print("Example usage: recommend_products('85123A', 5)")
```
